Получаем node-параметры для координат моего дома и университета, их мы и будем потом использовать как точки начала и конца пути

```python
import osmnx as ox

place = "Moscow, Russia"
G = ox.graph_from_place(place, network_type="drive")

hous_lat, hous_lon = 55.650892, 37.726181
uni_lat, uni_lon = 55.710749, 37.603911

# fetch the nearest node w.r.t coordinates
hous_node = ox.distance.nearest_nodes(G, hous_lon, hous_lat)
uni_node = ox.distance.nearest_nodes(G, uni_lon, uni_lat)

print("House Node: {hous_node}, \
RUDN Node: {uni_node}".format(
    hous_node=hous_node, uni_node=uni_node))
```

Подставляем в функцию координаты и ищем кратчайший путь на авто по расстоянию

```python
import osmnx as ox

place = "Moscow, Russia"
G = ox.graph_from_place(place, network_type="drive")

# orgin node and destination node
orig, dest = 251073892, 10870541544

# find shortest path
route_nodes = ox.routing.shortest_path(G, orig, dest, weight="length")

# plot the shortes path
fig, ax = ox.plot_graph_route(G, route_nodes, route_color="r", 
                              route_linewidth=6, node_size=0)
```

Подставляем в функцию координаты и ищем кратчайший путь на авто по времени. Визуально он совпал с кратчайшим по расстоянию. Почему бы и нет.

```python
import osmnx as ox

place = "Moscow, Russia"
G = ox.graph_from_place(place, network_type="drive")

# impute speed on all edges missing data
G = ox.add_edge_speeds(G)

# calculate travel time (seconds) for all edges
G = ox.add_edge_travel_times(G)

# orgin node and destination node
orig, dest = 251073892, 10870541544

# find shortest path
route_nodes = ox.routing.shortest_path(G, orig, dest, weight="travel_time")

# plot the shortes path
fig, ax = ox.plot_graph_route(G, route_nodes, route_color="r", 
                              route_linewidth=6, node_size=0)
```
