```python
import pandas as pd

data = pd.read_csv('GameOfThrones.csv')
data
```

Удалим из него лишние стоблцы и выведем остаток

```python
del data['Type']
del data['book']
```

```python
data
```

Создаем граф из всей этой красоты и выводим его на экран

```python
import networkx as nx
import matplotlib.pyplot as plt

Graf = nx.from_pandas_edgelist(data, 'Source', 'Target', ['weight'])

pos = nx.spring_layout(Graf)

labels = nx.get_edge_attributes(Graf,'weight')


plt.figure(figsize=(10,6))


nx.draw_networkx(Graf,
        pos,
        with_labels=True,
        node_color='red',
        node_size=5,       
        width=0.1,         
        font_size=3)       
plt.show()
```

Мы видим, что множество вершин "на периферии" не связаны друг с другом напрямую, т.е. граф не связный.
